## Geocoding Places/villages using Nominatim API 

```python
for index, row in df.iterrows():
    text=str(row['Village/Locality name'])
    #bounding coordinates added to restrict search results to city of banglore( approx)
    client = requests.session()
    request = client.get(url='https://nominatim.openstreetmap.org/?format=json&addressdetails=1&q='+text+'&format=json&limit=1&countrycodes=in&bounded=1&viewbox=77.180737,12.732146,77.968935,13.504153')
    
    # variable x is of type string.the size of empty result is 2 ( ie x=[]) 
    x=request.text
    
    if len(x)!= 2:
        parsed_x = json.loads(x)
        df.iloc[index,2]=parsed_x[0]['lat']
        df.iloc[index,3]=parsed_x[0]['lon']
        df.iloc[index,4]=parsed_x[0]['boundingbox'][0]
        df.iloc[index,5]=parsed_x[0]['boundingbox'][1]
        df.iloc[index,6]=parsed_x[0]['boundingbox'][2]
        df.iloc[index,7]=parsed_x[0]['boundingbox'][3]
        
    time.sleep(1)
```

### CSV CONTAINS THE LAT ,LON AND BOUNDING BOX DETAILS OF EACH PINCODE

```python
dftest=pd.read_csv("test.csv")
dfres=pd.read_csv("results_places.csv")
```

```python
dftest=dftest.drop("Unnamed: 0",1)
dftest
```

```python
dfres=dfres.drop("Unnamed: 0",1)
dfres
```

filling in the null values using the result obtained previously by geocoding places/villages

```python
for index, row in dftest.iterrows():
  location=row['Village/Locality name']
  x=dfres.loc[dfres['Village/Locality name']== location]
  lat=x.iloc[0,2]
  lon=x.iloc[0,3]
  bblata=x.iloc[0,4]
  bblatb=x.iloc[0,5]
  bblona=x.iloc[0,6]
  bblonb=x.iloc[0,7]
  dftest.iloc[index,2]=lat
  dftest.iloc[index,3]=lon
  dftest.iloc[index,4]=bblata
  dftest.iloc[index,5]=bblatb
  dftest.iloc[index,6]=bblona
  dftest.iloc[index,7]=bblonb
```

#### Grouping the results based on Pincodes 
The idea is to obtain the Bounding Box coordinates primarly using the Pincode itself.The remaining slots can be filled out using the coordinates of the various places in the region .



1.   we select all the places having a certain pincode and group them together 
2.   we select the minimum latitude and longitude among the box coordinates as the south-west corner of the box 
3.  similarly we select the maximum lat&lon as the North - East corner of the Box 


---




##NOTICE

####     The above algorithm is not Fool proof as we cannot conclude that the  bounding box coordinates so obtained will cover the entire region of Pincode . 



```python
dfpin=dftest.groupby("Pincode")
```

```python
for name , group in dfpin:
  print (name)
  print (group)
```

```python
dffinal=dfpin.agg({'BB_lat_1': np.min,'BB_lat_2': np.max,'BB_lon_1': np.min,'BB_lon_2': np.max})
```

```python
dffinal
```

```python
dffinal.to_csv("result_places_pincode.csv")
```
